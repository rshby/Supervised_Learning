## Prepare Corpus

```python
df = pd.read_csv("datasets/spam_or_not_spam.csv")
df = df.loc[df["email"].notna()]
df.head()
```

```python
# membuat tokenize
sentences = [word_tokenize(email.lower()) for email in tqdm(df["email"])]
sentences[:5]
```
