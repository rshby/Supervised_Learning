```python
import pandas as pd
```

```python
# Retrieve HTML table data
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html = pd.read_html(url, header = 0)
df2019 = html[0]

# Data cleaning
raw = df2019.drop(df2019[df2019.Age == 'Age'].index)
raw
```

## Data Cleaning

```python
raw.shape
```

```python
# Cek data yang bolong
raw.isnull().sum()
```

## Mengganti missing value dengan angka 0

```python
df = raw.fillna(0)
df.head()
```

```python
# Cek data yang bolong lagi
df.isnull().sum()
```

## Hapus kolom RK

```python
df.drop(columns="Rk", inplace=True)
df.shape
```

## Menyimpan df menjadi csv

```python
df.to_csv("nba2019.csv", index=False)
```

## QUESTIONS
### Conditional Selection
In performing exploratory data analysis, it is important to be able to select subsets of data to perform analysis or comparisons.

Which player scored the most Points (PTS) Per Game? Here, we will return the entire row.

```python
# melihat data yang pts-nya paling tinggi(max)
filt = df["PTS"] == df["PTS"].max()
df.loc[filt]
```

Further question, what team is the player from?

```python
# melihat darimana team player yang memiliki pts paling tinggi
df.loc[filt][["Player", "Tm"]]
```

Which position is the player playing as?

```python
# melihat posisi plater yang memiliki PTS paling tinggi
df.loc[filt][["Player", "Pos"]]
```

How many games did the player played in the season?

```python
# melihat banyaknya pemain tersebut bermain di satu season
df.loc[filt][["Player", "G"]]
```

Which player scored more than 20 Points (PTS) Per Game?

```python
# melihat data yang memiliki pts diatas 20
df.loc[df["PTS"] > 20][["Player", "Pos", "Age", "Tm", "G", "PTS"]]
```

Which player had the highest 3-Point Field Goals Per Game (3P) ?

```python
# melihat pemain yang memiliki 3P paling tinggi
df.loc[df["3P"] == df["3P"].max()][["Player", "Pos", "Tm", "3P"]]
```

Which player had the highest Assists Per Game (AST) ?

```python
# melihat pemain yang meiliki nilai AST paling tinggi
df.loc[df["AST"] == df["AST"].max()]
```

Which player scored the highest (PTS) in the Los Angeles Lakers?

```python
LAL = df.groupby(by='Tm').get_group('LAL')
LAL.loc[LAL["PTS"] == LAL["PTS"].max()]
```
