# Mengganti nama kolom

```python
import pandas as pd
```

```python
df = pd.read_csv("https://bit.ly/uforeports")
df.head()
```

```python
# melihat kolom apa saja yang ada di df
df.columns
```

```python
# menggunakan cara dictionary
df.rename(columns={
    "Colors Reported" : "Colors_Reported"
}, inplace=True)
df.columns
```

```python
# mengganti nama kolom sesuai daftar array kita
kolom = ["city", "colors reported", "shape reported", "state", "time"]
df.columns = kolom
df.columns
```

```python
# parameter names=array_columns
df = pd.read_csv("https://bit.ly/uforeports", names=kolom, header=0)
df.head()
```

```python
# mengganti semua spasi(" ") dalam kolom dengan underscore("_")
df.columns = df.columns.str.replace(" ","_")
df.columns
```

# Menghapus kolom pada DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv("https://bit.ly/uforeports")
df.head()
```

```python
# dimensi DataFrame
df.shape
```

```python
# Menggunakan drop("column_name", axis=1, inplace=True)
df.drop("Colors Reported", axis=1, inplace=True)
df.head()
```

```python
# Menhapus lebih dari 1 kolom secara langsung
# menggunakan drop(["kolom", "kolom"])
df.drop(["City", "State"], axis=1, inplace=True)
df.head(3)
```

```python
# Menghapus baris
# drop([index], axis=0, inplace=True)
df.drop([0, 1], axis=0, inplace=True)
df.head(3)
```

# Filter baris berdasarkan syarat nilai kolom

```python
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/imdbratings")
df.head()
```

```python
# dimensi DataFrame
df.shape
```

```python
# filter data yang "duration" > 200
df.loc[df["duration"] >= 200]
```

```python
booleans = []
for lenght in df["duration"]:
    if lenght >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
        
booleans[:5]
```

```python
len(booleans)
```

```python
is_long = pd.Series(booleans)
is_long
```

```python
df[is_long]
```

```python
# atau menggunakan seperti ini
is_long = df["duration"] >= 200
is_long
```

```python
df[is_long]
```

```python
df.loc[is_long]
```

```python
# Menggunakan DataFrame[kondisi filter]
df[df["duration"]>= 200]
```

```python
# Filter data, namun hanya menampilkan kolom tertentu
# DataFrame[kondisi][["kolom", "kolom", ...]]
df[df["duration"]>=200][["title", "genre"]]
```

```python
#df.loc[kondisi, ["kolom", "kolom"]]
df.loc[df["duration"]>=200, ["title", "genre"]]
```

# Melakukan filter dengan beberapa kondisi

```python
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/imdbratings")
df.head()
```

```python
# 2 kondisi
# df[(kondisi) & (kondisi)]
df[(df["duration"]>=200) & (df["genre"]=="Drama")]
```

```python
# 3 kondisi
# df[(kondisi) & (kondisi) & (kondisi)]
df[(df["duration"]>=200) & (df["genre"]=="Drama") & (df["star_rating"]>8)]
```

```python
# menggunakan isin
df[df["genre"].isin(["Crime", "Drama"])]
```
