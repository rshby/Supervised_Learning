```python
import matplotlib.pyplot as plt
import numpy as np
```

# Membuat plotting sederhana
Membuat sebuah `figure` yang memiliki sebuah `axes`. Melakukan plotting pada `axes`.

```python
# membuat sample data
x = [1,2,3,4]
y = [1,4,2,3]

fig, ax = plt.subplots() # membuat figure dan sebuah axes
ax.plot(x, y)
plt.show()
```

Alternatif lain dapat langsung memanfaatkan method `plt.plot()` pada matplotlib untuk melakukan plotting sederhana

```python
x = [1,2,3,4]
y = [1,4,2,3]

plt.plot(x, y)
plt.show()
```

2. `pyplot` style

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots() # membuat sebuah figure dan sebuah axes.

#plotting 3 varian data pada axes
ax.plot(x, x, label="linear")
ax.plot(x, x**2, label="quadratic")
ax.plot(x, x**3, label="cubic")

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_tittle('x label')
```
